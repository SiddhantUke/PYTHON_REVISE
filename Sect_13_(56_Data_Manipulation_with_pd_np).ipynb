```python
## Data Aggregating  and Grouping !

df.head()
```

```python
## Product ----> Vased on this we are actually do the mean of this specific value !

grouped_mean=df.groupby('Product')['Value'].mean()
print(grouped_mean)
```

```python
## I want to do groupby based on the categorical variables !

## Product and Region !

grouped_sum=df.groupby(['Product','Region'])['Value'].sum()
print(grouped_sum)
```

```python

## Actucally a mean  of the upper "Product" and "Region"

df.groupby(['Product','Region'])['Value'].mean()
```

```python
## Aggregaet Multiple Functn !

## aggregate multiple functions


groudped_agg=df.groupby('Region')['Value'].agg(['mean','sum','count'])
groudped_agg
```

```python
### Merging and joining Dataframes
# Create sample DataFrames


df1 = pd.DataFrame({'Key': ['A', 'B', 'C'], 'Value1': [1, 2, 3]})
df2 = pd.DataFrame({'Key': ['A', 'B', 'D'], 'Value2': [4, 5, 6]})
```

```python
df1
```

```python
df2
```

```python
## Merge Dataframes on the 'Key Columns'

pd.merge(df1,df2,on="Key",home = 'inner')

## on="Key" --------> Basically means on which columns name !

## Outer -----------> Both dataframe is consider and probably each row and column is considered !

## Left Outer Join -------------> df1 will given more priority that is given in dataframe !

## RIght --------> Right  side key is more important
```

```python
pd.merge(df1,df2,on="Key",how="outer")
```

```python
pd.merge(df1,df2,on="Key",how="left")
```

```python
pd.merge(df1,df2,on="Key",how="right")
```
